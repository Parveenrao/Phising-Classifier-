```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("phishing.csv")
data
```

```python
data.info()
```

```python
plt.figure(figsize=(8,5))  # Optional: Adjusts figure size
sns.countplot(x=data['class'], palette='coolwarm')  # Adjust palette for colors
plt.title('Status Count Distribution', fontsize=14)  # Add title
plt.xlabel('Status', fontsize=12)  # Label x-axis
plt.ylabel('Count', fontsize=12)  # Label y-axis
plt.xticks(rotation=45)  # Rotate x-axis labels if needed
plt.show()  # Display the plot
```
